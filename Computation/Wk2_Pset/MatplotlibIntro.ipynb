```python
import numpy as np
from matplotlib import pyplot as plt
```

# Problem 1

```python
def varmatrix(n):
    A=np.eye(n)
    for i in range(0,n):
        for j in range(0,n):
            A[i,j]=np.random.normal()
    meanrow=np.mean(A,axis=0)
    var=np.var(meanrow)
    return var

N=np.ones(10)
for n in range(10):
    N[n]=varmatrix(100*(n+1))
graph=plt.plot(N)
plt.show(graph)
```

# Problem 4

```python
x=np.linspace(0,2*np.pi,100)

plt.subplot(321)
plt.plot(x, np.sin(x),'g-',lw=2)
plt.title("Sin(x)")
plt.ylim(-2,2)


plt.subplot(322)
plt.plot(x,np.sin(2*x),'r--')
plt.ylim(-2,2)
plt.title("Sin(2x)")


plt.subplot(325)
plt.plot(x,2*np.sin(x),'b--')
plt.ylim(-2,2)
plt.title("2sin(x)")


plt.subplot(326)
plt.plot(x,2*np.sin(2*x),'m:')
plt.ylim(-2,2)
plt.title("2sin(2x)")

plt.suptitle("Sinx and Sin2x")
plt.show()
```

# Problem 5

```python
fars=np.load('FARS.npy')
#[hour,longi,lati]

plt.subplot(131)
long=fars[:,1]
lat=fars[:,2]
plt.plot(long,lat,'ko', markersize=0.1)
plt.axis("equal")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Longitude vs Latitude")

plt.subplot(133)
plt.hist(fars[:,0],bins=24,range=[0,24])
plt.xlabel("Hour")
plt.ylabel("Number of accidents")
plt.show()
```

# Problem 6

```python
x=np.linspace(-2*np.pi,2*np.pi,100)
y=x.copy()
X,Y=np.meshgrid(x,y)
Z=(np.sin(X)*np.sin(Y))/(X*Y)

plt.subplot(121)
plt.pcolormesh(X,Y,Z, cmap='magma')
plt.colorbar()
plt.xlim(-2*np.pi,2*np.pi)
plt.ylim(-2*np.pi,2*np.pi)
plt.title("Heat map")

plt.subplot(122)
plt.contour(X,Y,Z,np.linspace(-0.9,0.9,30),cmap='Spectral')
plt.colorbar()
plt.xlim(-2*np.pi,2*np.pi)
plt.ylim(-2*np.pi,2*np.pi)
plt.title("Contour map")

plt.show()
```
