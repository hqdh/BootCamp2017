```python
import numpy as np
import pandas as pd
from datetime import datetime
from matplotlib import pyplot as plt
```

```python
payems=pd.read_csv('payems.csv', header=None).shift(-6)
payems=pd.DataFrame({'date': payems[0], 'payems': payems[1]}).set_index('date')
payems.index=pd.to_datetime(payems.index)
payems=payems.dropna()
for i in range(len(payems['payems'])):
    payems.iloc[i,:]=1000*int(payems.iloc[i,:])
new_year1=pd.Series(payems['2029':'2099'].index).apply(lambda x: x - pd.DateOffset(years=100))
new_year2=pd.Series(payems[:'2017'].index)
realdate=new_year1.append(new_year2)
payems.index=realdate
#create segments
p1=payems['1928-08':'1940-03']
p2=payems['1936-05':'1945-06']
p3=payems['1944-02':'1952-10']
p4=payems['1947-11':'1956-10']
p5=payems['1952-07':'1961-05']
p6=payems['1956-08':'1965-04']
p7=payems['1959-04':'1968-02']
p8=payems['1968-12':'1977-11']
p9=payems['1972-11':'1982-03']
p10=payems['1979-01':'1987-07']
p11=payems['1980-07':'1989-11']
p12=payems['1989-07':'1998-03']
p13=payems['2000-03':'2008-11']
p14=payems['2006-12':'2016-06']
#normalize:
p1=p1/p1.iloc[0,:]
p2=p2/p2.iloc[1,:]
p3=p3/p3.iloc[12,:]
p4=p4/p4.iloc[12,:]
p5=p5/p5.iloc[12,:]
p6=p6/p6.iloc[12,:]
p7=p7/p7.iloc[12,:]
p8=p8/p8.iloc[12,:]
p9=p9/p9.iloc[12,:]
p10=p10/p10.iloc[12,:]
p11=p11/p11.iloc[12,:]
p12=p12/p12.iloc[12,:]
p13=p13/p13.iloc[12,:]
p14=p14/p14.iloc[12,:]
#plot:
x1=np.linspace(1,8,len(p1))
x2=np.linspace(0,8,len(p2))
x3=np.linspace(0,8,len(p3))
x4=np.linspace(0,8,len(p4))
x5=np.linspace(0,8,len(p5))
x6=np.linspace(0,8,len(p6))
x7=np.linspace(0,8,len(p7))
x8=np.linspace(0,8,len(p8))
x9=np.linspace(0,8,len(p9))
x10=np.linspace(0,8,len(p10))
x11=np.linspace(0,8,len(p11))
x12=np.linspace(0,8,len(p12))
x13=np.linspace(0,8,len(p13))
x14=np.linspace(0,8,len(p14))
plt.plot(x1,p1, 'k-', lw=5, label='1928-1940')
plt.plot(x2,p2, label='1936-1945')
plt.plot(x3,p3, label='1944-1952')
plt.plot(x4,p4, label='1947-1956')
plt.plot(x5,p5, label='1952-1961')
plt.plot(x6,p6, label='1956-1965')
plt.plot(x7,p7, label='1959-1968')
plt.plot(x8,p8, label='1968-1977')
plt.plot(x9,p9, label='1972-1982')
plt.plot(x10,p10, label='1979-1987')
plt.plot(x11,p11, label='1980-1989')
plt.plot(x12,p12, label='1989-1998')
plt.plot(x13,p13, label='2000-2008')
plt.plot(x14,p14, 'r-', lw=5, label='2006-2016')
plt.xlabel('Time from peak')
plt.ylabel('Jobs/peak')

plt.legend()
plt.axhline(y=1, color='gray', linestyle='--')
plt.axvline(x=1, color='gray', linestyle='--')
x=np.linspace(0,8,9)
myxticks=['-1yr','peak','+1yr','+2yr','+3yr','+4yr','+5yr','+6yr','+7yr']
plt.xticks(x, myxticks)
plt.savefig('recession.png')
plt.show()

```
