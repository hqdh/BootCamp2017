# Problem 2

```python
def bernstein(n,v,x):
    if n<v:
        raise ValueError("n has to be larger than or equal to v")
    return scipy.special.binom(n,v)*(x**v)*(1-x)**(n-v)

x=np.linspace(0,1,100)

for n in range(4):
    for v in range(n+1):
        plt.subplot(4,4,n+4*v+1)
        plt.plot(x,bernstein(n,v,x))
        plt.ylim=(0,1)
plt.suptitle("10 first Bernsteins")
plt
plt.show()
```

# Problem 5

```python
from matplotlib.colors import LogNorm
x=np.linspace(-3,3,800)
X,Y=np.meshgrid(x,x.copy())
Z=(1-X)**2+100*(Y-X**2)**2

plt.subplot(121)
plt.pcolormesh(X,Y,Z,cmap='viridis',norm=LogNorm())
plt.colorbar()

plt.subplot(122)
plt.contour(X,Y,Z,cmap='viridis',norm=LogNorm())
plt.colorbar()
plt.axis("equal")
plt.show()
```

# Problem 6

```python
data=np.load('countries.npy')
countries = ["Austria", "Bolivia", "Brazil", "China",
"Finland", "Germany", "Hungary", "India",
"Japan", "North Korea", "Montenegro", "Norway",
"Peru", "South Korea", "Sri Lanka", "Switzerland",
"Turkey", "United Kingdom", "United States", "Vietnam"]
popu=data[:,0]
GDP=data[:,1]
maleheight=data[:,2]
femheight=data[:,3]

plt.subplot(221)
plt.bar(np.arange(len(countries)), GDP, align="center")
plt.xticks(np.arange(len(countries)), countries)
plt.xlabel("Countries")
plt.xlabel("Countries and GDP")
plt.ylabel("GDP")

plt.subplot(222)
plt.hist(popu,bins=len(countries))
plt.xlabel("Population")
plt.ylabel("Number of countries")

plt.subplot(223)
plt.scatter(maleheight,femheight,s=5)
plt.title("Average Male height vs Average Female height")
plt.xlabel("Average Male height")
plt.ylabel("Average Female height")

plt.subplot(224)
plt.scatter(popu,GDP,s=5)
plt.title("Population vs GDP")
plt.xlabel("Population")
plt.ylabel("GDP")

plt.show()

#Average male height and average female height are strongly correlated. In general, population and GDP are also 
#correlated, but it may not be the case for countries with large population.
```
