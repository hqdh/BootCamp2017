```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

# Problem 7

```python
lipids=pd.read_csv('lipids.csv', header=None)
lipids=lipids.shift(-5)
lipids['diseased']=lipids[0]
lipids['chol']=lipids[1]
lipids['trig']=lipids[2]
lipids=lipids[['diseased','chol','trig']].dropna()
t=np.array(lipids[lipids['diseased']=='1']['chol'])
s=np.array(lipids[lipids['diseased']=='1']['trig'])
chol=np.array(list(map(int,t)))
trig=np.array(list(map(int,s)))
plt.hist(chol, bins=25)
plt.savefig('2Dhist.png')
plt.show()
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
hist, xedges, yedges=np.histogram2d(chol, trig, bins=[25,25])
xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25)
xpos = xpos.flatten('F')
ypos = ypos.flatten('F')
zpos = np.zeros_like(xpos)
dx = 25**np.ones_like(zpos)
dy = dx.copy()
dz=hist.flatten()
ax.bar3d(xpos, ypos, zpos, dx, dy, dz, color='b', zsort='average')
plt.savefig('3Dhist.png')
plt.show()
```
