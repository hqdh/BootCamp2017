```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
chic=pd.read_csv('Chic.csv')
ind=pd.read_csv('Indi.csv')
mia=pd.read_csv('Miam.csv')
wash=pd.read_csv('Wash.csv')
pitt=pd.read_csv('Pitt.csv')
chic['DATE']=pd.to_datetime(chic['DATE'], format='%Y%m%d')
ind['DATE']=pd.to_datetime(ind['DATE'], format='%Y%m%d')
mia['DATE']=pd.to_datetime(mia['DATE'], format='%Y%m%d')
wash['DATE']=pd.to_datetime(wash['DATE'], format='%Y%m%d')
pitt['DATE']=pd.to_datetime(pitt['DATE'], format='%Y%m%d')
chic=chic.set_index('DATE')[['TMAX','TMIN']]
ind=ind.set_index('DATE')[['TMAX','TMIN']]
mia=mia.set_index('DATE')[['TMAX','TMIN']]
wash=wash.set_index('DATE')[['TMAX','TMIN']]
pitt=pitt.set_index('DATE')[['TMAX','TMIN']]
s=pd.DataFrame(chic.index.dayofyear)
x=pd.date_range(start='1999-09-21', end='2000-09-20', freq='D')
chic.index=(chic.index.dayofyear-263)%366
ind.index=(ind.index.dayofyear-263)%366
mia.index=(mia.index.dayofyear)%366
wash.index=(wash.index.dayofyear)%366
pitt.index=(pitt.index.dayofyear)%366
plt.scatter(chic.index, chic['TMAX'], label='Chicago', c='maroon', s=0.5)
plt.scatter(pitt.index, pitt['TMAX'], label='Pittsburgh', c='green', s=0.5)
plt.scatter(ind.index, ind['TMAX'], label='Indianapolis', c='black', s=0.5)
plt.scatter(mia.index, mia['TMAX'], label='Miami', c='yellow', s=0.5)
plt.scatter(wash.index, wash['TMAX'], label='Washington', c='blue', s=0.5)
plt.xlabel('Day of year')
plt.ylabel('Temperature')
plt.legend()
plt.scatter(wash.index, wash['TMAX'], c='blue', s=0.5)
plt.scatter(mia.index, mia['TMAX'], c='yellow', s=0.5)
plt.scatter(pitt.index, pitt['TMIN'], c='green', s=0.5)
plt.scatter(ind.index, ind['TMAX'],c='black', s=0.5)
plt.scatter(chic.index, chic['TMIN'], c='maroon', s=0.5)
plt.savefig('temperature.png')
plt.show()
```
